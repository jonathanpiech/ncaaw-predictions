```python
reg_det = pd.read_csv('WDataFiles_Stage1/WRegularSeasonDetailedResults.csv')
reg_det
```

```python
list(reg_det)
```

```python
reg_det.reset_index(inplace=True)
reg_det.rename(columns={'index': 'GameID'}, inplace=True)
reg_det.head()
```

```python
reg_win = reg_det.drop(['LTeamID',
 'LScore',
 'LFGM',
 'LFGA',
 'LFGM3',
 'LFGA3',
 'LFTM',
 'LFTA',
 'LOR',
 'LDR',
 'LAst',
 'LTO',
 'LStl',
 'LBlk',
 'LPF'], axis=1)
```

```python
reg_win.rename(columns={'WTeamID': 'TeamID', 'WScore': 'Score', 'WLoc': 'Loc', 'WFGM': 'FGM', 'WFGA': 'FGA', 'WFGM3': 'FGM3', 'WFGA3': 'FGA3', 'WFTM': 'FTM', 'WFTA': 'FTA', 'WOR': 'OR', 'WDR': 'DR', 'WAst': 'Ast', 'WTO': 'TO', 'WStl': 'Stl', 'WBlk': 'Blk', 'WPF': 'PF'}, inplace=True)
```

```python
reg_win['Result'] = 1
```

```python
reg_lose = reg_det.drop(['WTeamID',
 'WScore',
 'WFGM',
 'WFGA',
 'WFGM3',
 'WFGA3',
 'WFTM',
 'WFTA',
 'WOR',
 'WDR',
 'WAst',
 'WTO',
 'WStl',
 'WBlk',
 'WPF',
 ], axis=1)
```

```python
reg_lose.rename(columns={'LTeamID': 'TeamID', 'LScore': 'Score', 'WLoc': 'Loc', 'LFGM': 'FGM', 'LFGA': 'FGA', 'LFGM3': 'FGM3',
                         'LFGA3': 'FGA3', 'LFTM': 'FTM', 'LFTA': 'FTA', 'LOR': 'OR', 'LDR': 'DR', 'LAst': 'Ast', 'LTO': 'TO',
                         'LStl': 'Stl', 'LBlk': 'Blk', 'LPF': 'PF'}, inplace=True)
```

```python
reg_lose
```

```python
def loser_location(loc):
    if loc == 'H':
        return 'A'
    elif loc == 'A':
        return 'H'
    else:
        return loc
```

```python
reg_lose['Loc'] = reg_lose['Loc'].apply(loser_location)
```

```python
reg_lose['Result'] = 0
```

```python
reg = pd.concat((reg_win, reg_lose)).reset_index(drop=True)
```

```python
reg
```

```python
#Possessions = FGA-OR+TO+.475*FTA
reg['Poss'] = reg['FGA'] - reg['OR'] + reg['TO'] + 0.475 * reg['FTA']
```

```python
reg.sort_values(by='Poss')
```
