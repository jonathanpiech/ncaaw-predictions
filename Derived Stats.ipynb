# Derive Stats to use in Model

```python
#Import dependencies
import pandas as pd
import numpy as np
import matplotlib as plt
```

## Separate game data from winning Team vs. losing Team to team vs. opponent

```python
#Load dataframe of regular season game stats from 2010-2019
reg_det = pd.read_csv('WDataFiles_Stage1/WRegularSeasonDetailedResults.csv')
reg_det
```

```python
#Create GameID number
reg_det.reset_index(inplace=True)
reg_det.rename(columns={'index': 'GameID'}, inplace=True)
reg_det.head()
```

```python
#Create copy of dataframe for winning teams
reg_win = reg_det.copy()
reg_win
```

```python
#Rename columns: winning team stats to team stats and losing team stats to opposing team stats
reg_win.rename(columns={'WTeamID': 'TeamID', 'WScore': 'Score', 'WLoc': 'Loc', 'WFGM': 'FGM', 'WFGA': 'FGA',
                        'WFGM3': 'FGM3', 'WFGA3': 'FGA3', 'WFTM': 'FTM', 'WFTA': 'FTA', 'WOR': 'OR', 'WDR': 'DR',
                        'WAst': 'Ast', 'WTO': 'TO', 'WStl': 'Stl', 'WBlk': 'Blk', 'WPF': 'PF', 'LTeamID': 'OTeamID',
                        'LScore': 'OScore', 'LFGM': 'OFGM', 'LFGA': 'OFGA', 'LFGM3': 'OFGM3', 'LFGA3': 'OFGA3',
                        'LFTM': 'OFTM', 'LFTA': 'OFTA', 'LOR': 'OOR', 'LDR': 'ODR', 'LAst': 'OAst', 'LTO': 'OTO',
                        'LStl': 'OStl', 'LBlk': 'OBlk', 'LPF': 'OPF'}, inplace=True)
```

```python
#Create copy of dataframe for losing teams
reg_lose = reg_det.copy()

#Rename columns: losing team stats to team stats and winning team stats to opposing team stats
reg_lose.rename(columns={'WTeamID': 'OTeamID', 'WScore': 'OScore', 'WLoc': 'Loc', 'WFGM': 'OFGM', 'WFGA': 'OFGA',
                        'WFGM3': 'OFGM3', 'WFGA3': 'OFGA3', 'WFTM': 'OFTM', 'WFTA': 'OFTA', 'WOR': 'OOR', 'WDR': 'ODR',
                        'WAst': 'OAst', 'WTO': 'OTO', 'WStl': 'OStl', 'WBlk': 'OBlk', 'WPF': 'OPF', 'LTeamID': 'TeamID',
                        'LScore': 'Score', 'LFGM': 'FGM', 'LFGA': 'FGA', 'LFGM3': 'FGM3', 'LFGA3': 'FGA3',
                        'LFTM': 'FTM', 'LFTA': 'FTA', 'LOR': 'OR', 'LDR': 'DR', 'LAst': 'Ast', 'LTO': 'TO',
                        'LStl': 'Stl', 'LBlk': 'Blk', 'LPF': 'PF'}, inplace=True)
```

```python
#Create function to swap Home and Away for losing team dataframes
def loser_location(loc):
    if loc == 'H':
        return 'A'
    elif loc == 'A':
        return 'H'
    else:
        return loc
```

```python
#Swap Home and Away
reg_lose['Loc'] = reg_lose['Loc'].apply(loser_location)
```

```python
#Add result column to winning and losing team dataframes
reg_win['Result'] = 1
reg_lose['Result'] = 0
```

```python
#Reorder columns of winning teams
reg_win = reg_win[['GameID',
 'Season',
 'DayNum',
 'Loc',
 'NumOT',
 'TeamID',
 'Score',
 'FGM',
 'FGA',
 'FGM3',
 'FGA3',
 'FTM',
 'FTA',
 'OR',
 'DR',
 'Ast',
 'TO',
 'Stl',
 'Blk',
 'PF',
 'OTeamID',
 'OScore',
 'OFGM',
 'OFGA',
 'OFGM3',
 'OFGA3',
 'OFTM',
 'OFTA',
 'OOR',
 'ODR',
 'OAst',
 'OTO',
 'OStl',
 'OBlk',
 'OPF',
 'Result']]
```

```python
#Reorder columns of losing teams
reg_lose = reg_lose[['GameID',
 'Season',
 'DayNum',
 'Loc',
 'NumOT',
 'TeamID',
 'Score',
 'FGM',
 'FGA',
 'FGM3',
 'FGA3',
 'FTM',
 'FTA',
 'OR',
 'DR',
 'Ast',
 'TO',
 'Stl',
 'Blk',
 'PF',
 'OTeamID',
 'OScore',
 'OFGM',
 'OFGA',
 'OFGM3',
 'OFGA3',
 'OFTM',
 'OFTA',
 'OOR',
 'ODR',
 'OAst',
 'OTO',
 'OStl',
 'OBlk',
 'OPF',
 'Result']]
```

```python
#Combine winning and losing team dataframes
reg = pd.concat((reg_win, reg_lose)).reset_index(drop=True)
```

## Use regular season data to derive stats

```python
#Create possession and pace stats

#Possessions = FGA-OR+TO+.475*FTA
reg['Poss'] = reg['FGA'] - reg['OR'] + reg['TO'] + 0.475 * reg['FTA']
reg['OPoss'] = reg['OFGA'] - reg['OOR'] + reg['OTO'] + 0.475 * reg['OFTA']

#Pace is the average amount of possessions of the game between both teams
reg['Pace'] = (reg['OPoss'] + reg['Poss']) / 2
```

```python
#Create offensive and defensive ratings for the team per game

#Offensive Rating = Score/Pace * 100
#Defensive Rating = Opposing Score/Pace * 100
reg['OFF'] = reg['Score'] / reg['Pace'] * 100
reg['DEF'] = reg['OScore'] / reg['Pace'] * 100
```

```python
#Group by season and team id to get a team's average offensive rating for the season
off_rating = reg.groupby(['Season', 'TeamID']).mean()
```

```python
#Select only the offensive rating column
off_rating = off_rating[['OFF']]
```

```python
#Rename column to reflect the team's average offensive rating for the season
off_rating.rename(columns={'OFF': 'AvgOFF'}, inplace=True)
```

```python
#Reset index
off_rating.reset_index(inplace=True)
```

```python
#Merge team's average offensive rating to regular season game data
reg = pd.merge(reg, off_rating, how='left', left_on=['Season', 'TeamID'], right_on=['Season', 'TeamID'])
```

```python
#Select all games that did not go to overtime to get accurate number of possessions per game (pace)
pace_rating = reg[reg.NumOT == 0]

#Group by season and team id to get a team's average pace rating for the season
pace_rating = reg.groupby(['Season', 'TeamID']).mean()
pace_rating = pace_rating[['Pace']]
pace_rating.rename(columns={'Pace': 'AvgPace'}, inplace=True)
pace_rating.reset_index(inplace=True)
reg = pd.merge(reg, pace_rating, how='left', left_on=['Season', 'TeamID'], right_on=['Season', 'TeamID'])
```

```python
#Group by season and opposing team id to get a team's average defensive rating for the season
def_rating = reg.groupby(['Season', 'OTeamID']).mean()
def_rating = def_rating[['DEF']]
def_rating.rename(columns={'DEF': 'AvgDEF'}, inplace=True)
def_rating.reset_index(inplace=True)
reg = pd.merge(reg, def_rating, how='left', left_on=['Season', 'OTeamID'], right_on=['Season', 'OTeamID'])
```

```python
#Group by season and team id to get a team's number of wins for the season
wins = reg.groupby(['Season', 'TeamID']).sum()
wins = wins[['Result']]
wins.rename(columns={'Result': 'Wins'}, inplace=True)
wins.reset_index(inplace=True)
```

```python
#Group by season and team id to get a team's number of games for the season
games = reg.groupby(['Season', 'TeamID']).count()
games = games[['Result']]
games.rename(columns={'Result': 'Games'}, inplace=True)
games.reset_index(inplace=True)
```

```python
#Merge wins and games dataframe on season and team id
win_games = pd.merge(wins, games, how='left', left_on=['Season', 'TeamID'], right_on=['Season', 'TeamID'])
```

```python
#Calculate winning percentage = wins/games
win_games['WinP'] = win_games['Wins'] / win_games['Games']
```

```python
#Merge winning percentage dataframe with regular season game dataframe
reg_add = pd.merge(reg, win_games, how='left', left_on=['Season', 'TeamID'], right_on=['Season', 'TeamID'])
```

```python
#Create a dataframe for opposing team's winning percentage

#Drops wins and games columns kepping winning percentage column
opp_win = win_games.drop(['Wins', 'Games'], axis=1)

#Rename columns from team to opponent
opp_win.rename(columns={'WinP': 'OWinP', 'TeamID': 'OTeamID'}, inplace=True)
```

```python
#Merge opponents winning percentage onto regular season game dataframe
reg_add_opp = pd.merge(reg_add, opp_win, how='left', left_on=['Season', 'OTeamID'], right_on=['Season', 'OTeamID'])
```

```python
#Group by season and team id to get a team's average winning percentage of the opposing team
avg_opp = reg_add_opp.groupby(['Season', 'TeamID']).mean()
avg_opp = avg_opp[['OWinP']]
avg_opp.reset_index(inplace=True)
```

```python
#Drops the column with the winning percentage of the opposing team
reg_add_opp.drop(['OWinP'], axis=1, inplace=True)

#Merge data frame with the average winning percentage of the opposing team
reg_avg_opp = pd.merge(reg_add, avg_opp, how='left', left_on=['Season', 'TeamID'], right_on=['Season', 'TeamID'])
```

```python
#Group by season and team id to get a team's average opponent's winning percentage of opponents
avg_opp_opp = reg_avg_opp.groupby(['Season', 'TeamID']).mean()
avg_opp_opp = avg_opp_opp[['OWinP']]
avg_opp_opp.rename(columns={'OWinP': 'OppOWinP'}, inplace=True)
avg_opp_opp.reset_index(inplace=True)
```

```python
#Merge opponent's opponents winning percentage to regular season dataframe
reg_rpi = pd.merge(reg_avg_opp, avg_opp_opp, how='left', left_on=['Season', 'TeamID'], right_on=['Season', 'TeamID'])
```

```python
#Calculate RPI = .25 * winning percentage + .5 * opponent's winning percentage + .25 opponent's opponent's winning percentage
reg_rpi['RPI'] = 0.25*reg_rpi['WinP'] + 0.5*reg_rpi['OWinP'] + 0.25*reg_rpi['OppOWinP']
reg_rpi
```

```python
#Save to csv
reg_rpi.to_csv("derived_stats.csv", index=False)
```
